### Problem 7. YouTube views and likes (1 point)
What is the relationship between YouTube views and likes? Present an appropriate plot. 

What is the mean YouTube views-to-likes ratio? What is its distribution? Find a way to plot it and describe it.

Ensure the columns are numeric and handle missing values:

```python
spotify_df['youtube_views'] = pd.to_numeric(spotify_df['youtube_views'].str.replace(',', ''), errors='coerce')
spotify_df['youtube_likes'] = pd.to_numeric(spotify_df['youtube_likes'].str.replace(',', ''), errors='coerce')
```

Drop rows with missing YouTube views or likes:

```python
youtube_df = spotify_df.dropna(subset=['youtube_views', 'youtube_likes'])
```

Scatter plot of YouTube views and likes:

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='youtube_views', y='youtube_likes', data=youtube_df)
plt.xlabel('YouTube Views')
plt.ylabel('YouTube Likes')
plt.title('Relationship Between YouTube Views and Likes')
plt.show()
```

Calculating the YouTube views-to-likes ratio:

```python
youtube_df['views_to_likes_ratio'] = youtube_df['youtube_views'] / youtube_df['youtube_likes']
```

Mean YouTube views-to-likes ratio:

```python
mean_ratio = youtube_df['views_to_likes_ratio'].mean()
print(f"Mean YouTube views-to-likes ratio: {mean_ratio}")
```

Distribution of the YouTube views-to-likes ratio:

```python
plt.figure(figsize=(10, 6))
sns.histplot(youtube_df['views_to_likes_ratio'], bins=30, kde=True)
plt.xlabel('YouTube Views-to-Likes Ratio')
plt.title('Distribution of YouTube Views-to-Likes Ratio')
plt.show()
```

### Problem 8. TikTok stuff (2 points)
The most popular songs on TikTok released every year show... interesting behavior. Which years peaked the most TikTok views? Show an appropriate chart. Can you explain this behavior? For a bit of context, TikTok was created in 2016.

Now, how much popular is the most popular song for each release year, than the mean popularity? Analyze the results.

In both parts, it would be helpful to see the actual songs.

Ensure TikTok views are numeric:

```python
spotify_df['tiktok_views'] = pd.to_numeric(spotify_df['tiktok_views'].str.replace(',', ''), errors='coerce')
```

Determine the years with the highest TikTok views

```python
tiktok_views_by_year = spotify_df.groupby('release_date')['tiktok_views'].sum().reset_index()
tiktok_views_by_year['release_date'] = pd.to_datetime(tiktok_views_by_year['release_date']).dt.year
```

Summarize TikTok views by year:

```python
tiktok_views_by_year = tiktok_views_by_year.groupby('release_date')['tiktok_views'].sum().reset_index()
```

Plot total TikTok views by year:

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='release_date', y='tiktok_views', data=tiktok_views_by_year, palette='viridis')
plt.xlabel('Year')
plt.ylabel('Total TikTok Views')
plt.title('Total TikTok Views by Year')
plt.show()
```

Compare the most popular song to the mean popularity for each release year:

```python
spotify_df['release_year'] = pd.to_datetime(spotify_df['release_date']).dt.year
mean_tiktok_views_by_year = spotify_df.groupby('release_year')['tiktok_views'].mean().reset_index()
max_tiktok_views_by_year = spotify_df.loc[spotify_df.groupby('release_year')['tiktok_views'].idxmax()]
```

Merge the mean and max dataframes:

```python
comparison_df = pd.merge(mean_tiktok_views_by_year, max_tiktok_views_by_year[['release_year', 'track', 'artist', 'tiktok_views']], on='release_year')
comparison_df.columns = ['release_year', 'mean_tiktok_views', 'track', 'artist', 'max_tiktok_views']
comparison_df['popularity_ratio'] = comparison_df['max_tiktok_views'] / comparison_df['mean_tiktok_views']
```

Display the comparison:

```python
comparison_df = comparison_df.sort_values(by='release_year')
print(comparison_df[['release_year', 'track', 'artist', 'max_tiktok_views', 'mean_tiktok_views', 'popularity_ratio']])
```

Plot the popularity ratio:

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='release_year', y='popularity_ratio', data=comparison_df, palette='viridis')
plt.xlabel('Year')
plt.ylabel('Popularity Ratio')
plt.title('Most Popular Song\'s TikTok Views to Mean TikTok Views Ratio by Year')
plt.show()
```
