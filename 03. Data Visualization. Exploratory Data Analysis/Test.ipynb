#### Power Plant Cost
__The "Nuclear" dataset contains information about the construction of several nuclear power plants across the US in the 1960s and 1970s. The dataset is located at https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/boot/nuclear.csv. The data description file is at https://stat.ethz.ch/R-manual/R-devel/library/boot/html/nuclear.html. Read the data description to get some understanding of what's in the data. Load the dataset into pandas. You can either download it, or use the URL directly. The following questions are related to this dataset. What is the mean cost of a power plant, in millions of USD? Round your answer to 2 decimal points.__

```python
# Load the dataset
url = "https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/boot/nuclear.csv"
data = pd.read_csv(url)

# Display the first few rows of the dataset to understand its structure
print(data.head())
```

```python
data.columns
```

```python
data
```

```python
# Calculate the mean cost
mean_cost = data['cost'].mean()

# Round to 2 decimal points
mean_cost_rounded = round(mean_cost, 2)

print(f"The mean cost of a power plant is ${mean_cost_rounded} million.")
```

Answear: 461.56

#### Number of Stations
__How many of all stations were constructed in the north-east region of the US?__

```python
# Filter the rows where the 'ne' column has a value of 1
north_east_stations = data[data['ne'] == 1]

# Count the number of such rows
num_north_east_stations = len(north_east_stations)

print(f"Number of stations constructed in the north-east region of the US: {num_north_east_stations}")
```

Answear: 8

#### Cost over Time
__What is the general behavior of the cost of a power plant over time? Write "increasing", "decreasing" or "constant", without the quotes.__

```python
data.dtypes
```

```python
#Convert 'date' to integer type (if not already)
data['date'] = data['date'].astype(int)

# Sort the data by 'date'
data_sorted = data.sort_values(by='date')

# Plot the cost over time
plt.figure(figsize=(10, 6))
plt.plot(data_sorted['date'], data_sorted['cost'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Cost (in millions of USD)')
plt.title('Cost of Power Plants Over Time')
plt.grid(True)
plt.show()
```

Answear: increasing

#### Max Power
__Some stations are built on places where there were existing stations in the past. Out of all of these, how much power does the most powerful one generate, in MWe? Your answer should be a whole number.__

```python
# Load the dataset
url = "https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/boot/nuclear.csv"
data = pd.read_csv(url)

# Inspect the dataset
print(data.head())
print(data.columns)

# Filter the dataset to include only stations built at locations with existing stations in the past
existing_stations = data[data['cum.n'] > 0]

# Find the maximum power generated by these stations
max_power = existing_stations['cap'].max()

# Print the result
print(f"The maximum power generated by the most powerful station at an existing location is: {int(max_power)} MWe")
```
